```python
df['SibSp'].value_counts()
```

```python
df['Individual_fare'] = df['Fare']/(df['SibSp'] + df['Parch'] + 1)
```

```python
df['Individual_fare'].value_counts()
```

```python
df['Individual_fare'].plot(kind = 'box')
```

```python
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
```

```python
df
```

```python
def transfer_family_size(num):
    if num == 1:
        return 'Alone'
    if num <=4:
        return 'Small'
    else:
        return 'Large'
```

```python
df['Family_type']=df['Family_Size'].apply(transfer_family_size)
df
```

```python
pd.crosstab(df['Survived'],df['Family_Size'],normalize='columns')*100
```
